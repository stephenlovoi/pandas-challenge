```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data_df["SN"].value_counts())
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_unique = len(purchase_data_df["Item Name"].unique())
average_price = round(purchase_data_df["Price"].mean(), 2)
number_of_purchases = len(purchase_data_df["Purchase ID"].value_counts())
total_revenue = purchase_data_df["Price"].sum()
item_summary_df = pd.DataFrame({"Number of Unique Items": [item_unique], 
                           "Average Price": f"${average_price}", 
                           "Number of Purchases": [number_of_purchases], 
                           "Total Revenue": f"${total_revenue}"})
item_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_df = purchase_data_df.drop_duplicates(subset=["SN"])
gender_count = gender_df["Gender"].value_counts()
gender_count_df = pd.DataFrame(gender_count)
gender_percentage = round((gender_count_df["Gender"] / gender_count_df["Gender"].sum()) * 100, 2).map("{:.2f}%".format)
gender_count_df["Percentage of Players"] = gender_percentage

#rename 'Gender' column to 'Total Count'
gender_count_df = gender_count_df.rename(columns = {"Gender": "Total Count"})

gender_count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_count = purchase_data_df["Gender"].value_counts()
gender_analysis_df = pd.DataFrame(gender_purchase_count)

# ** NOT WORKING** rename 'Gender' Column to "Total Count" for cleaner format
gender_analyis_df = gender_analysis_df.rename(columns = {"Gender": "Purchase Count"})

gender_analysis_df
```

```python
# finding the average purchase price and total purchase value of each gender
male_purchase_data = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
male_purchase_average = round(male_purchase_data["Price"].mean(), 2)
male_purhcase_total = male_purchase_data["Price"].sum()

female_purchase_data = purchase_data_df.loc[purchase_data_df["Gender"] == "Female",:]
female_purchase_average = round(female_purchase_data["Price"].mean(), 2)
female_purhcase_total = female_purchase_data["Price"].sum()

other_purchase_data = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]
other_purchase_average = round(other_purchase_data["Price"].mean(), 2)
other_purhcase_total = other_purchase_data["Price"].sum()

# create lists that will be columns in the final output
average_purchase_price_list = [male_purchase_average, female_purchase_average, other_purchase_average]
total_purchase_value_list = [male_purhcase_total, female_purhcase_total, other_purhcase_total]

# final output
gender_analysis_df["Average Purchase Price"] = average_purchase_price_list
gender_analysis_df["Total Purchase Value"] = total_purchase_value_list

#add new column 'Avg. Total Purchase per Person
gender_analysis_df["Avg. Total Purchase per Person"] = gender_analysis_df["Total Purchase Value"] / gender_count_df["Total Count"]

# mapping for cleaner data
gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].map("${:.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map("${:.2f}".format)
gender_analysis_df["Avg. Total Purchase per Person"] = gender_analysis_df["Avg. Total Purchase per Person"].map("${:.2f}".format)

#display data
gender_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#making bins and labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#slicing data and put it into bins
pd.cut(purchase_data_df["Age"], bins, labels = age_labels)
```

```python
# place the data series into a new column inside of the DataFrame
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels = age_labels)

#drop repeated names so we do not over/underestimate a gender
sn_drop = purchase_data_df.drop_duplicates(subset = ["SN"])

# group by Age Group
age_group = sn_drop.groupby("Age Group")

age_group_df = age_group[["Age"]].count()
age_group_df["Percentage of Players"] = round(age_group_df["Age"] / age_group_df["Age"].sum() * 100, 2).map("{:.2f}%".format)

# rename "Age" column to "Total Count" for a cleaner display
age_group_df = age_group_df.rename(columns = {"Age": "Total Count"})
age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create new dataframe and group by "Age Group" Column
purchase_group = purchase_data_df.groupby("Age Group")["Price"].describe()[["count", "mean"]]
purchase_group = purchase_group.rename(columns = {"count": "Purchase Count", "mean": "Average Purchase Price"})

# add new columns
purchase_group["Total Purchase Value"] = purchase_group["Purchase Count"] * purchase_group["Average Purchase Price"]
purchase_group["Avg. Total Purchase per Person"] = purchase_group["Total Purchase Value"] / age_group_df["Total Count"]

# format for cleaner data
purchase_group["Purchase Count"] = purchase_group["Purchase Count"].map("{:.0f}".format)
purchase_group["Average Purchase Price"] = purchase_group["Average Purchase Price"].map("${:.2f}".format)
purchase_group["Total Purchase Value"] = purchase_group["Total Purchase Value"].map("${:.2f}".format)
purchase_group["Avg. Total Purchase per Person"] = purchase_group["Avg. Total Purchase per Person"].map("${:.2f}".format)

#display data
purchase_group_df = pd.DataFrame(purchase_group)
purchase_group_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#make new dataframe
spender_group = purchase_data_df.groupby("SN")["Price"].describe()[["count", "mean"]]
spender_group_df = pd.DataFrame(spender_group)
spender_group_df = spender_group_df.rename(columns = {"count": "Purchase Count", "mean": "Average Purchase Price"})

#add new column
spender_group_df["Total Purchase Value"] = spender_group_df["Purchase Count"] * spender_group_df["Average Purchase Price"]

#sort total purchase value in descending order
spender_group_df = spender_group_df.sort_values("Total Purchase Value", ascending = False)

#format for cleaner data
spender_group_df["Purchase Count"] = spender_group_df["Purchase Count"].map("{:.0f}".format)
spender_group_df["Average Purchase Price"] = spender_group_df["Average Purchase Price"].map("${:.2f}".format)
spender_group_df["Total Purchase Value"] = spender_group_df["Total Purchase Value"].map("${:.2f}".format)

spender_group_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create a new dataframe (NOTE: can keep "mean" because the item price remains the same -- mean of item price = item price)
item_group_df = purchase_data_df.groupby(["Item ID", "Item Name"])["Price"].describe()[["count", "mean"]]
item_group_df = item_group_df.rename(columns = {"count": "Purchase Count", "mean": "Item Price"})
item_group_df["Total Purchase Value"] = item_group_df["Purchase Count"] * item_group_df["Item Price"]

# filling
item_group_df["Purchase Count"] = item_group_df["Purchase Count"].fillna(0)

#sort so purchase count is in descening order
item_group_df = item_group_df.sort_values("Purchase Count", ascending = False)

#format data
item_group_df["Purchase Count"] = item_group_df["Purchase Count"].map("{:.0f}".format)
item_group_df["Item Price"] = item_group_df["Item Price"].map("${:.2f}".format)
item_group_df["Total Purchase Value"] = item_group_df["Total Purchase Value"].map("${:.2f}".format)

item_group_df.head()
```
